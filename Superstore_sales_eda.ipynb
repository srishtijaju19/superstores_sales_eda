```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('superstore.csv', encoding='latin1')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
#check for null
df.isnull().sum()
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
```

#Univariate Analysis

```python
# Sales Disttibution
sns.histplot(df['Sales'], bins = 50)
```

```python
# Proit Distribution
sns.histplot(df['Profit'], bins = 50)
```

```python
#Category Counts
df['Category'].value_counts().plot(kind='bar')
```

#Bivariate Analyis

```python
# Profit by Category
sns.boxplot(x='Category', y='Profit', data = df)
```

```python
# Sales & Profit by Region
region_sales = df.groupby('Region')[['Sales', 'Profit']].sum().sort_values('Sales', ascending=False)
region_sales.plot(kind='bar', title='Sales & Profit by Region')
```

```python
# Profit by Subcategory
#sns.boxplot(x='Sub-Category', y='Profit', data=df)

subcat = df.groupby('Sub-Category')['Profit'].sum().sort_values('Profit', ascending=False)
subcat.plot(kind='bar', title='Profit by Subcategory')
```

```python
#Ship Mode
sns.countplot(x='Ship Mode', data = df)
```

# Time Series Trends

```python
df['Month'] = df['Order Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')[['Sales', 'Profit']].sum()

monthly_sales.plot(figsize=(12,5), title='Monthly Sales & Profit')
```

### Key Insight
- West if the most profitable category, whereas south is struggling
- Tables as a subcategory has consistently been loss making
- December sees spike in sales but not in profit which could means poor discounting offers

### Recommendations
- Re-evalaute pricing strategy for unprofitable sub-categories
- Investigate shipping costs in the South region.
- Consider promoting highly profitable categories more during high-sale months.
